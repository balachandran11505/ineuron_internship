```python
import pandas as pd
import numpy as np
```

```python
#CPI Excel file
cpi_mon=pd.read_excel('CPI Data.xlsx',sheet_name='monthly')
cpi_ann=pd.read_excel('CPI Data.xlsx',sheet_name='annual')

#Exchange rate file
exchange_mon=pd.read_excel('Exchange rate.xlsx',sheet_name='monthly')
exchange_ann=pd.read_excel('Exchange rate.xlsx',sheet_name='annual')

#Exports Merchandise file
exports_mon=pd.read_excel('Exports Merchandise.xlsx',sheet_name='monthly')
exports_quart=pd.read_excel('Exports Merchandise.xlsx',sheet_name='quarterly')
exports_ann=pd.read_excel('Exports Merchandise.xlsx',sheet_name='annual')
```

```python
cpi_annual = cpi_ann.iloc[1:]
```

```python
cpi_annual.rename(columns = {'Unnamed: 0':'Year'}, inplace = True)
cpi_annual['Year'] = cpi_annual.Year.astype('int64').astype('string')
cpi_annual=cpi_annual.fillna(0)
```

```python
cpi_mon.rename(columns = {'Unnamed: 0':'Year_Month'}, inplace = True)
```

```python
def get_quarter(month):
    if month in ['01','02','03']:
        return "Q1"
    elif month in ['04','05','06']:
        return "Q2"
    elif month in ['07','08','09']:
        return "Q3"
    elif month in ['10','11','12']:
        return "Q4"
    else:
        return "Invalid Month"
```

```python
month_dict = {
    '01': "January",
    '02': "February",
    '03': "March",
    '04': "April",
    '05': "May",
    '06': "June",
    '07': "July",
    '08': "August",
    '09': "September",
    '10': "October",
    '11': "November",
    '12': "December"
}
```

```python
year_mon = []
year_qrt = []
year = []
for i in cpi_mon['Year_Month']:
    year.append(i[0:4])
    year_mon.append(month_dict[i[5:]])
    year_qrt.append(get_quarter(i[5:]))
cpi_mon['Year_Month'] = year_mon
cpi_mon['Year'] = year
cpi_mon['Quarter'] = year_qrt
cpi_mon.rename(columns = {'Year_Month':'Month'}, inplace = True)
```

```python
cpi_mon=cpi_mon.fillna(0)
```
